# Final Reasons For The Varying Revenue Across US States :


## 1. Unit Price Is Not the Key Driver
- Unit prices are relatively stable across states, with only minor fluctuations (~$100â€“200).  
- Therefore, **revenue differences are not caused by price variations**.

## 2. Order Quantity Drives Revenue
- Total revenue is highly correlated with **total order quantity**.  
- States with higher order quantities naturally generate higher revenue.

## 3. Market Size Matters
- Larger states like **California, Texas, and Florida** have bigger markets, more customers, and more orders.  
- California, in particular, shows the **highest number of orders and the largest total order quantity**, resulting in the **highest revenue**.

## 4. Channel and Product Mix Amplify Differences
- Wholesale orders dominate in high-revenue states, contributing to large bulk sales.  
- Top products (like : Product 26 ) are concentrated in large states, further boosting total revenue.  
- Smaller states have fewer customers, smaller order quantities, and limited product mix.

## 5. Smaller States Naturally Show Low Revenue
- States with smaller populations and markets (e.g., Maine, Mississippi, Delaware) have **fewer orders, lower order quantities, and fewer customers**, resulting in significantly lower revenue.

---

## Conclusion
The primary factor driving revenue differences across US states is **order quantity**, which itself is influenced by **market size, number of customers, and channel/product distribution**.  
Unit price plays a minor role, while high-volume wholesale orders and concentration of top products in large states amplify revenue differences.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Assuming your dataframe is named df
# Features for clustering
X = df[['revenue', 'profit_margin_pct']]

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Run K-Means with 3 clusters (you can tune n_clusters)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Cluster centers (scaled â†’ transform back to original scale)
centers = scaler.inverse_transform(kmeans.cluster_centers_)

# Plot clusters
plt.figure(figsize=(10,6))
plt.scatter(df['revenue'], df['profit_margin_pct'], c=df['Cluster'], cmap='viridis', alpha=0.6)
plt.scatter(centers[:,0], centers[:,1], c='red', marker='X', s=200, label='Centroids')
plt.title("Customer Segmentation: Revenue vs Profit Margin (KMeans Clusters)")
plt.xlabel("Revenue")
plt.ylabel("Profit Margin (%)")
plt.legend()
plt.show()

# Analyze clusters
cluster_summary = df.groupby('Cluster').agg({
    'revenue':'mean',
    'profit_margin_pct':'mean',
    'customer names':'nunique',
    'channel':'value_counts'
})


```

```python
df['margin'] = df['revenue'] - df['total_cost']
df['margin_pct'] = df['margin'] / df['revenue']
```

```python
sku_summary = df.groupby('product_name').agg({
    'order_quantity':'sum',
    'revenue':'sum',
    'margin':'sum',
    'margin_pct':'mean'   # average margin %
}).reset_index()
```

```python
sku_summary.sort_values("margin",ascending=False)
```

```python
# Contribution of each SKU
sku_summary['revenue_share'] = sku_summary['revenue'] / sku_summary['revenue'].sum()

# Sort by revenue descending
sku_summary = sku_summary.sort_values('revenue', ascending=False)

# Cumulative revenue share (Pareto)
sku_summary['cum_revenue_share'] = sku_summary['revenue_share'].cumsum()

# Mark if SKU is in top 80% revenue
sku_summary['pareto_flag'] = sku_summary['cum_revenue_share'].apply(lambda x: 'Core (Top 80%)' if x <= 0.8 else 'Long Tail (Bottom 20%)')
```

```python
# Low sales = bottom quartile
sales_threshold = sku_summary['order_quantity'].quantile(0.25)
```

```python
def classify(row):
    if row['margin_pct'] >= 0.2 and row['pareto_flag'] == 'Core (Top 80%)':
        return "Retain"
    elif row['margin_pct'] < 0.2 and row['pareto_flag'] == 'Core (Top 80%)':
        return "Improve"
    elif row['order_quantity'] < sales_threshold or row['pareto_flag'] == 'Long Tail':
        return "Rationalize"
    else:
        return "Retain"

sku_summary['strategy'] = sku_summary.apply(classify, axis=1)
```

```python


# Load data
# Example for Product 1

def predict_per_sku(sku):
    sku_df = df[df['product_name'] == sku].groupby('order_month')['order_quantity'].sum().reset_index()
    
    # Convert 'order_month' to datetime if not already
    # Convert Period dtype to Timestamp
    sku_df['order_month'] = sku_df['order_month'].dt.to_timestamp()
    
    
    # Prophet requires columns 'ds' (date) and 'y' (value to predict)
    sku_df.rename(columns={'order_month': 'ds', 'order_quantity': 'y'}, inplace=True)
    
    # Build model
    model = Prophet(yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
    model.fit(sku_df)
    
    # Forecast next 6 months
    future = model.make_future_dataframe(periods=12, freq='M')
    forecast = model.predict(future)
    
    # Plot forecast
    model.plot(forecast)
    plt.show()
```
